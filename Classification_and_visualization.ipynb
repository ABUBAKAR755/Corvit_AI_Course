```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
full_data = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
print('train data:' , full_data.shape)
```

```python
full_data.head(5)
```

```python
full_data.info()
```

```python
sns.heatmap(full_data.isnull(), yticklabels = False, cbar=False,cmap='tab20c_r')
plt.title('Missing Data: Training Set')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass', y='Age', data=full_data,palette='GnBu_d')
plt.show()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age

full_data['Age'] = full_data[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
full_data.drop('Cabin', axis=1, inplace=True)
```

```python
full_data.dropna(inplace=True)
```

```python
full_data.drop(['Name' , 'Ticket'], axis=1, inplace=True)
objcat = ['Sex', 'Embarked']
for colname in objcat:
    full_data[colname] = full_data[colname].astype('category')
    full_data[colname] = full_data[colname].cat.codes
```

```python
full_data.describe()
```

```python
full_data.drop('PassengerId', inplace=True, axis=1)
```

```python
full_data.shape
```

```python
full_data.select_dtypes(['category']).columns

#Index(['Sex', 'Embarked'], dtype='object')
```

```python
sex = pd.get_dummies(full_data['Sex'],drop_first=True)
embark = pd.get_dummies(full_data['Embarked'],drop_first=True)
```

```python
full_data.head()
```

```python
full_data = pd.concat([full_data,sex,embark],axis=1)
full_data.head(5)
```

```python
full_data.drop(['Sex','Embarked'],axis=1,inplace=True)
print('train_data shape',full_data.shape)
full_data.head()
```

```python
x = full_data.drop('Survived', axis=1)
y = full_data['Survived']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= .20 , random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

# Convert column names to strings to avoid TypeError
x_train.columns = x_train.columns.astype(str)

# Create and train the model
Ireg = LogisticRegression()
Ireg.fit(x_train, y_train)
```

```python
x_test.columns = x_test.columns.astype(str)
y_pred_Ireg = Ireg.predict(x_test)
print(y_pred_Ireg)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print('Classification Model')
print('--' * 30)

logreg_accuracy = round(accuracy_score(y_test, y_pred_Ireg) * 100, 2)
print('Accuracy:', logreg_accuracy)

print(classification_report(y_test, y_pred_Ireg))
```
